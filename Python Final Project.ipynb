```python
#(4) From airport_freqs, extract all communication frequencies used by each large airport in New York State. 
# Sort the frequencies in ascending order.

# Buffalo Niagara International Airport 

buffalo_freq = airport_frequencies[(airport_frequencies['airport_ident'] == 'KBUF')] 
buffalo_freq.sort_values(by= 'frequency_mhz', ascending = True)
```

```python
# John F Kennedy International Airport Frequencies

John_Kennedy_freq = airport_frequencies[(airport_frequencies['airport_ident'] == 'KJFK')] 
John_Kennedy_freq.sort_values(by= 'frequency_mhz', ascending = True)
```

```python
#La Guardia Airport Frequencies

LaGuardia_freq = airport_frequencies[(airport_frequencies['airport_ident'] == 'KLGA')] 
LaGuardia_freq.sort_values(by= 'frequency_mhz', ascending = True)
```

```python
# Syracuse Hancock International Airport Frequencies

Syracuse_freq = airport_frequencies[(airport_frequencies['airport_ident'] == 'KSYR')] 
Syracuse_freq.sort_values(by= 'frequency_mhz', ascending = True)
```

```python
# III. Airports over the world

# (1)Create a new data frame large_airports by selecting rows from airports related to large airports.

large_airports = airports[(airports['type'] == 'large_airport')] 
large_airports
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# (2) Use the groupby mechanic to split large_airports by countries.
group = large_airports['type'].groupby(large_airports['iso_country'])

group
```

```python
#. (3) Calculate the number of large airports in each country.

group = large_airports['type'].groupby(large_airports['iso_country']).count()

print(group.to_string())

#Below, I don't know if the ipynb file or pdf will show it but 
# setting the group.to_string() prints out the calculated
# number of large airports in each country.
```

```python
#4. Find the top 5 countries having the most large airports. 
# Display the full country names in your result.

group = large_airports['type'].groupby(large_airports['iso_country']).count()

top_countries = group.sort_values(ascending = False)

print("The top 5 countries with the most large airports are: The United States, China, Mexico, Russia, and Italy \n", top_countries.head(5))
```
